```python
import numpy as np

#Each slot machine's winning probability, uniform distribution between 0-1
probs = np.random.uniform(size=10) # 10 probabilities for 10 slot machines

#Record rewards for each slot machine
rewards = [[0] for _ in range(10)] 

probs, rewards
```

```python
import random


#Greedy algorithm
def choose_one():
    #Small probability to randomly select a lever
    if random.random() < 0: 
        return random.randint(0, 9)

    #Calculate average reward for each slot machine
    rewards_mean = [np.mean(i) for i in rewards]

    #Select the lever with highest expected reward
    return np.argmax(rewards_mean)


choose_one()
```

```python
def try_and_play():
    i = choose_one()

    #Play the slot machine and get result
    reward = 0
    if random.random() < probs[i]:
        reward = 1

    #Record the result
    rewards[i].append(reward)


try_and_play()

rewards
```

```python
def get_result():
    #Play N times
    n=1000
    for _ in range(n):
        try_and_play()

    #Best possible expected reward
    target = probs.max() * n

    #Actual accumulated reward
    result = sum([sum(i) for i in rewards])

    return target, result


get_result()
```
