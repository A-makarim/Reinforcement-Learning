```python
import numpy as np

# Each slot machine's winning probability, uniform distribution between 0-1
probs = np.random.uniform(size=10)

# Record rewards for each slot machine
rewards = [[] for _ in range(10)]

probs, rewards
```

```python
import random


# Greedy algorithm with decreasing probability of random selection
def choose_one():
    # Calculate how many times the game has been played TO BE FILLED
    played_count = sum(len(i) for i in rewards)

    # Probability of random selection decreases over time

    if played_count == 0: # Avoid division by zero
        return random.randint(0, 9) # Randomly select a lever for the first play
    
    if random.random() < 1/played_count: # base on 1/n. Approaches zero
        return random.randint(0, 9) # random exploration takes place

    # Calculate average reward for each slot machine
    rewards_mean = [np.mean(i) for i in rewards]

    # Choose the arm with highest estimated reward
    return np.argmax(rewards_mean)
```

```python
choose_one()
```

```python
def try_and_play():
    i = choose_one() # choose random or max (after average reward is calculated)

    # Play the slot machine and get result
    reward = 0
    if random.random() < probs[i]:
        reward = 1

    # Record the result
    rewards[i].append(reward)
```

```python
try_and_play()
rewards
```

```python
def get_result():
    # Play N times
    n = 1000
    for _ in range(n):
        try_and_play()

    # Best possible expected reward
    target = probs.max() * n

    # Actual accumulated reward
    result = sum([sum(i) for i in rewards])

    return target, result
```

```python
get_result()
```
