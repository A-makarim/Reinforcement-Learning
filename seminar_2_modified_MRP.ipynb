```python
import numpy as np

# we have 6 states. our P will be 6 by 6

P = [[0.9, 0.1, 0,0,0,0 ],
     [0.5,0, 0.5, 0,0,0],
     [0, 0, 0, 0.6, 0, 0.4],
     [ 0 , 0, 0 , 0 , 0.3, 0.7],
     [0, 0.2, 0.3, 0.5, 0, 0],
     [0,0, 0, 0, 0, 1]]

R = [-1, -2, -2, +10, +1, 0]

P = np.array(P)
R = np.array(R)

print(P.shape)
```

```python
gamma = 0.5

def compute_return(start_index, chain, gamma):
    G = 0
    for i in reversed(range(start_index, len(chain))): # ofc start is zero
        # we go back

        G = R[chain[i]] + gamma * G   # for a specific state and path


    return G


chain = [1, 2, 3, 6]   # chain is a list
chain = [x-1 for x in chain]

start_index = 0
G = compute_return(start_index, chain, gamma)

print(G)
```

```python
#bellman equation for MRP is V = R + gamma * P * V
#rearranging gives us (I - gamma * P) * V = R

def compute(P, R, gamma, states_num):
    rewards = np.array(R)
    rewards = rewards.reshape((-1, 1))
    value = np.linalg.inv(np.eye(states_num) - gamma * P).dot(rewards)
    return value

V = compute(P,R, gamma , 6)
V

#  V[state 1] is better for “what happens on average from state 1” when transitions are stochastic, not fixed to one path.






```
